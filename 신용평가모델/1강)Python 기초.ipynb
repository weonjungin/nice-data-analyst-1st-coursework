## arithmetic-and-comparison-operations

```python
int_var = 30
int_var = 0
int_var = -12
```

```python
float_var = 1234.567
float_var = -59.12
float_var = 1e-5
```

```python
oct_var = 0o13
hex_var = 0x2f
print(oct_var)
print(hex_var)
```

```python
a = True
b = False
print(a and b)
print(a or b)
```

```python
# 계산에 쓰일 변수를 정의
x, y = 2, 4
```

```python
print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
print("x // y =", x // y)
```

```python
print("x ** y =", x ** y)
print("x % y =", x % y)
```

```python
print("x == 2:", x == 2)
print("x != 1:", x != 1)
print("x > y:", x > y)
print("x <= y:", x <= y)
```

## string

```python
s1 = 'example string'
print(s1)
```

```python
s2 = "double quotes"
print(s2)
```

```python
s3 = '''This is
a multi-line string'''
print(s3)
```

```python
s4 = ("This is an\n"
    "implicitly joined\n"
    "string")
print(s4)
```

```python
x, y = "abc", "def"
```

```python
print("x + y =", x + y)
print("x * 3 =", x * 3)
print("x[1] =", x[1])
print("x[2:] =", x[2:])
print("x[-1] =", x[-1])
print("y[1:3] =", y[1:3])
# 아래는 잘못된 연산입니다.
# x[1] = 'd'
```

```python
x = "this is a string"
```

```python
print("x.find(\"s\") =", x.find("s"))
print("x.find(\"z\") =", x.find("z"))
print("x.upper() =", x.upper())
print("x.split() =", x.split())
print("x.replace(\"this\", \"that\") =", x.replace("this", "that"))
```
