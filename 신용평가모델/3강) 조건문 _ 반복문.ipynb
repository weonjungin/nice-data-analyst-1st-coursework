## conditional-statement

```python
x = 2
if x > 2:
    print("Greater than 2.")
elif x == 2:
    print("Equals to 2.")
else:
    print("Smaller than 2.")
```

```python
x = "a"
if x == "a" or x == "b":
    if x == "a":
        print("A")
    else:
        print("B")
```

```python
# 조건부 표현식
score = 80
passed = True if score >= 60 else False
print(passed)
```

```python
# 퀴즈 정답
# 주민등록번호를 받아 성별을 출력(1, 3: 남자, 2, 4: 여자)
id = "800101-2345678"

if id[7] == "1" or id[7] == "3":
# if id[7] in ("1", "3"):
    print("남자")
elif id[7] == "2" or id[7] == "4":
    print("여자")
else:
    print("주민등록번호가 잘못되었습니다.")
```

## iteration

```python
# for
l = [1, 2]
for elem in l:
    print(elem)
```

```python
sum = 0
for i in range(0, 5):
    sum += i
print(sum)
```

```python
# 인덱스 포함해서 리스트 반복
l = ["a", "b", "c"]
for idx, elem in enumerate(l):
    print(f"Index: {idx}, Value: {elem}")
```

```python
# 딕셔너리 반복
d = {1: "A", 2: "B", 3: "C"}
for k in d.keys():
    print(k, ":", d[k])

# 혹은 아래와 같이
# for k, v in d.items():
#     print(k, ":", v)
```

```python
# list comprehension
l1 = [x for x in range(1, 4)]
l1
```

```python
l2 = [x for x in range(1, 5) if x % 2 == 0]
l2
```

```python
# continue
for i in range(5):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# break
for i in range(5):
    if i >= 2:
        break
    print(i)
```

```python
# while
cnt = 0
while cnt < 2:
    cnt += 1
    print(cnt)
```

```python
cnt = 0
while True:
    cnt += 1
    print(cnt)
    if cnt >= 2:
        break
```

```python
# 퀴즈
# 문자열 리스트를 입력으로 받아 5글자 이상인 단어만 출력
input_list = ["apple", "banana", "cat", "dog", "elephant"]

for word in input_list:
    if len(word) >= 5:
        print(word)
```

```python
# 퀴즈 - 2
# 딕셔너리(키: 문자열, 값: 정수)를 입력으로 받아 값이 홀수인 키만 리스트로 생성
input_dict = {"A": 3, "B": 2, "C": 19, "D": 32, "E": 55}

# 정답 - 1
odds = []
for k, v in input_dict.items():
    if v % 2 != 0:
        odds.append(k)
print(odds)

# 정답 - 2
odds_2 = [k for k, v in input_dict.items() if v % 2 != 0]
print(odds_2)
```
