```python
# 재귀
# 재귀가 없는 일반 형태
def factorial_norec(n):
    answer = 1
    for i in range(2, n + 1):
        answer *= i
    return answer

factorial_norec(5)
```

```python
# 재귀 형태
def factorial_rec(n):
    # 종료 조건
    if n <= 1:
        print("End of the recursion...")
        return 1
    print(f"Calling facotiral_rec({n})")
    return n * factorial_rec(n - 1)

factorial_rec(5)
```

```python
# 두 숫자형을 입력으로 받아 최대값을 반환하는 함수 작성
def my_max(a, b):
    if a > b:
        return a
    else:
        return b

# 혹은...
# def my_max(a, b):
#     return a if a > b else b

# 혹은 2...
# def my_max(a, b):
#     return max(a, b)

my_max(3, 5)
```

```python
# 세 숫자형을 입력으로 받아 최대값을 반환하는 함수 작성
def my_max2(a, b, c):
    max = a
    if max < b:
        max = b
    if max < c:
        max = c

    return max

# 혹은...
# def my_max2(a, b, c):
#     return max(a, b, c)

my_max2(7, 5, 8)
```

## class

```python
# 스택 클래스 정의하기
class Stack:
    def __init__(self):
        self.l = []

    # __len__ 함수를 정의하면 len 함수에 객체를 인자로 전달할 수 있습니다.
    def __len__(self):
        return len(self.l)

    def push(self, new_elem):
        self.l.append(new_elem)

    def pop(self):
        popped = self.l[-1]
        del self.l[-1]
        return popped

    def top(self):
        return self.l[-1]

    def is_empty(self):
        return len(self.l) == 0
```

```python
s = Stack()
print("Stack empty:", s.is_empty())
s.push(1)
s.push(2)
print("Top of the stack:", s.top())
print("Stack size:", len(s))
s.pop()
s.pop()
print("Stack empty:", s.is_empty())
```
