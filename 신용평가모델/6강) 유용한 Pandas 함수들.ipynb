## pandas-funcs-3

```python
import pandas as pd
```

```python
# 샘플 데이터 - 1990년 California 주택 데이터
csv_file_url = "https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv"
df = pd.read_csv(csv_file_url, sep=",")
```

```python
# 사칙연산
df["non_bedrooms"] = df["total_rooms"] - df["total_bedrooms"]
df["rooms_per_households"] = df["total_rooms"] / df["households"]

df.loc[:, ["non_bedrooms", "rooms_per_households"]]
```

```python
# isna
df_with_na = pd.DataFrame({"A": [None, 1], "B": [2, None]})
df_with_na.isna()
```

```python
# fillna
df_with_na.fillna(0)
```

```python
# isin
df["housing_median_age"].isin([3.0, 4.0]).sum()
```

## numpy-func

```python
import numpy as np
```

```python
# arange
arr = np.arange(1, 6)
arr
```

```python
# zeros
arr = np.zeros((2, 2))
arr
```

```python
# full
arr = np.full((2, 2), 1)
arr
```

```python
# eye
arr = np.eye(2)
arr
```

```python
# binary operations
arr_1 = np.array([[2, 3], [1, 4]])
arr_2 = np.array([[-1, 2], [4, 3]])
```

```python
np.add(arr_1, arr_2)
```

```python
np.subtract(arr_1, arr_2)
```

```python
np.matmul(arr_1, arr_2)
```

```python
# 통계 함수들
arr = np.array([[2, 3, 7], [1, 4, 5]])
print(f"arr.max() = {arr.max()}")
print(f"arr.max(axis=0) = {arr.max(axis=0)}")
print(f"arr.max(axis=1) = {arr.max(axis=1)}")
print(f"arr.argmax(axis=1) = {arr.argmax(axis=1)}")
```

```python
print(f"arr.min(axis=0) = {arr.min(axis=0)}")
print(f"arr.argmin(axis=0) = {arr.argmin(axis=0)}")
```

```python
print(f"arr.sum() = {arr.sum()}")
print(f"arr.sum(axis=0) = {arr.sum(axis=0)}")
print(f"arr.mean(axis=1) = {arr.mean(axis=1)}")
```

```python
# Shaping operations
arr = np.array([[2, 3, 1], [1, 2, -1]])
arr.shape
```

```python
arr_reshaped = arr.reshape((3, 2))
arr_reshaped
```

```python
arr_flattened = arr.flatten()
arr_flattened
```

```python
arr_transposed = arr.transpose()
arr_transposed
```

```python
# Numerical operations
arr = np.array([2.2, 3.0, 1.3, 4.2])

print(f"np.exp(arr) = {np.exp(arr)}")
print(f"np.log(arr) = {np.log(arr)}")
print(f"np.log2(arr) = {np.log2(arr)}")
print(f"np.sqrt(arr) = {np.sqrt(arr)}")
print(f"np.power(arr, 3) = {np.power(arr, 3)}")
print(f"np.ceil(arr) = {np.ceil(arr)}")
print(f"np.floor(arr) = {np.floor(arr)}")
```
