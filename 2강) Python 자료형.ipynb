## list

```python
# 리스트 정의하기
l = []
l = [1, 2, 3]
l = ["a", 1, 6]
l = [[1, 2], [3, 4]]
```

```python
# 리스트 연산
l1, l2 = [1, 2, 3], [4, 5, 6]
```

```python
print("l1 + l2 =", l1 + l2)
print("l1 * 2 =", l1 * 2)
```

```python
print("l1[0] =", l1[0])
print("l1[-1] =", l1[-1])
print("l1[1:3] =", l1[1:3])
print("len(l1) =", len(l1))
```

```python
l1[2] = 1
print(l1)
```

```python
del l2[0]
print(l2)
```

```python
# 리스트 관련 함수
l = [6, 2, 3]
```

```python
l.append(1)
print(l)
```

```python
l.sort()
print(l)
# 혹은 sorted(l)로 리스트를 변경하지 않고 정렬된 리스트만 가져올 수 있음
print("sorted(l) =", sorted(l))
print("sorted(l, reverse=True) =", sorted(l, reverse=True)) # 내림차순
print(l)
```

```python
l.reverse()
print(l)
```

```python
l.insert(0, 2) # 0번째 index에 2를 삽입한다.
print(l)
```

```python
l.remove(2) # 가장 첫번째로 나오는 2를 삭제(index가 아니다!)
print(l)
```

```python
l.extend([4, 5])
print(l)
```

## tuple-dict

```python
# 튜플
t = (1, 2, "a")
t = (3, 4)
```

```python
# 튜플 연산
t1, t2 = (1, 2, 3), (4, 5)
```

```python
print("t1 + t2 =", t1 + t2)
print("t1 * 2 =", t1 * 2)
print("len(t1) =", len(t1))
# 아래는 잘못된 연산입니다.
# t1[0] = 1
# del t1[1]
```

```python
# 튜플 - 리스트 상호 변환
t = (1, 2, 3)
l = ["a", "b", "c"]
print(list(t))
print(tuple(l))
```

```python
# 딕셔너리
d = {"a": 1, "b": 2, "c": 3}
d = {"a": [1, 2], "b": "s", 1: 2}
```

```python
# 딕셔너리 연산
d = {"x": 1, "y": 2, 3: "c"}
print("d[\"x\"] =", d["x"])
print("d[3] =", d[3])
print("\"x\" in d =", "x" in d)
```

```python
# 딕셔너리 관련 함수
d = {1: "a", 2: "b"}
print("d.keys() =", d.keys())
print("d.values() =", d.values())
print("d.items() =", d.items())
```

## set

```python
# 집합 정의
s = {1, 2, 3}
s = set(["a", "b"])
s = set((3, 4, 5))
```

```python
# 집합 연산
s1, s2 = {1, 2, 3}, {2, 3, 4}
```

```python
print("s1 & s2 =", s1 & s2)
print("s1.intersection(s2) =", s1.intersection(s2))
print("s1 | s2 =", s1 | s2)
print("s1.union(s2) =", s1.union(s2))
print("s1 - s2 =", s1 - s2)
print("s1.difference(s2) =", s1.difference(s2))
```

```python
# 집합 관련 함수
s = set(["a", "b", "c"])
```

```python
s.add("d")
s
```

```python
s.update(["b", "e"])
s
```

```python
s.remove("c")
s
```
