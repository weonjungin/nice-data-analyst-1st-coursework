## pandas-funcs-1

```python
import pandas as pd
```

```python
# 샘플 데이터 - 1990년 California 주택 데이터
csv_file_url = "https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv"
df = pd.read_csv(csv_file_url, sep=",")
df
```

```python
# to_csv
df.to_csv("./my-housing-data.csv", sep="|", index=False)
!ls
```

```python
# head
df.head(n=2)
```

```python
# iloc
df.iloc[:2]
```

```python
df.iloc[[1, 2], [0, 1]]
```

```python
df.iloc[:, 1:3]
```

```python
# loc
df.loc[0, "median_house_value"]
```

```python
# boolean 인덱싱
mask = df["housing_median_age"] < 17
mask
```

```python
print(df.loc[mask, :])
```

```python
# 여러 개의 조건 설정
and_mask = (df["housing_median_age"] < 17) \
            & (df["median_house_value"] < 70000)
df.loc[and_mask, :]
```

```python
or_mask = (df["total_rooms"] > 6000) | (df["total_bedrooms"] > 1500)
df.loc[or_mask, :]
```
